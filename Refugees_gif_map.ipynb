```python
from datawrapper import Datawrapper
dw = Datawrapper(access_token = os.getenv("DATAWRAPPER_ACCESS_TOKEN"))
```

```python
import pandas as pd
```

```python
data = pd.read_csv("gif_map_long.csv")
```

```python
data['year'].unique()
```

```python
# For each year
for year in data['year'].unique():
    if pd.notna(year):
       
    
    # prep data
        mask_year = data['year'] == year
        dw_data = data[mask_year]
        # publish chart
        year_chart = dw.create_chart(title = f"Refugees from Afghanistan by state in {year}", chart_type = "d3-maps-choropleth", data = dw_data, folder_id=85466)
        dw.update_description(
            year_chart['id'],
            source_name = 'Dreher, A., Langlotz, S., Matzat, J., Parsons, C. and Mayda, A. (2020). Immigration, Political Ideologies and the Polarization of American Politics',
            source_url = 'https://www.refugeeresettlementdata.com/data.html',
            byline = 'Anastasia Valeeva',
        )
        properties = {
     "axes": {
          "values": "refugees",
          "keys": "state"
      },
 "visualize" : {
     "basemap": "us-states",
     "map-key-attr": "NAME"
      }
    }
        dw.update_metadata(year_chart['id'], properties)
        dw.publish_chart(year_chart['id'], display = False)
        filename = "images/"+str(year)+".png"
        dw.export_chart(year_chart['id'], width = 400, filepath=filename, display=False)
    else:
        print("Skipping! Null value.")
```

```python
for chart in dw.get_charts(limit = 60):
    if chart['title'][:25] == "Refugees from Afghanistan":
        dw.delete_chart(chart['id'])
    else:
        pass
```
